## HYPER

```python
from deabook import DEAweak
from deabook.constant import CET_ADDI, RTS_VRS1, RTS_VRS2,RTS_CRS,OPT_LOCAL
```

```python
def HYPER_CRS():

    model = DEAweak.DEAweak2(data,sent = "K L E=Y:CO2", gy=[1],gx=[0,0,0],gb=[1], rts=RTS_CRS, baseindex=None,refindex=None)
    res = model.optimize(solver="mosek")
    # print(model.info(1))
    return res

HYPER_CRS = HYPER_CRS()
HYPER_CRS
```

```python
def HYPER_VRS1():

    model = DEAweak.DEAweak2(data,sent = "K L E=Y:CO2", gy=[1],gx=[0,0,0],gb=[1], rts=RTS_VRS1, baseindex=None,refindex=None)
    res = model.optimize(solver="mosek")
    # print(model.info(1))
    return res

HYPER_VRS1=HYPER_VRS1()
HYPER_VRS1
```

```python
def HYPER_VRS2():

    model = DEAweak.DEAweak2(data,sent = "K L E=Y:CO2", gy=[1],gx=[0,0,0],gb=[1], rts=RTS_VRS2, baseindex=None,refindex=None)
    res = model.optimize(solver="mosek")
    # print(model.info(1))
    return res

HYPER_VRS2=HYPER_VRS2()
HYPER_VRS2
```

```python
HYPER_CRS = HYPER_CRS.drop(columns = ['optimization_status','rho'])

HYPER_VRS1 = HYPER_VRS1.drop(columns = ['optimization_status','rho'])
HYPER_VRS2 = HYPER_VRS2.drop(columns = ['optimization_status','rho'])


HYPER_CRS =  HYPER_CRS.rename(columns={'te':'MEE_CRS'})
HYPER_VRS1 =  HYPER_VRS1.rename(columns={'teuo':'MEE_BO_VRS_SAME', })
HYPER_VRS1 =  HYPER_VRS1.rename(columns={'teo':'MEE_O_VRS_SAME', })
HYPER_VRS2 =  HYPER_VRS2.rename(columns={'teuo':'MEE_BO_VRS_DIFF'})
HYPER_VRS2 =  HYPER_VRS2.rename(columns={'teo':'MEE_O_VRS_DIFF'})

HYPER_CRS = pd.concat([HYPER_CRS,yearid],axis=1)
HYPER_VRS1 = pd.concat([HYPER_VRS1,yearid],axis=1)
HYPER_VRS2 = pd.concat([HYPER_VRS2,yearid],axis=1)
```

```python
HYPER_CRS3 = HYPER_CRS.pivot_table(
    values=['MEE_CRS'],
    index=dmuname,
    columns='year',
    aggfunc='mean',
    fill_value=None,
    margins=False,
    dropna=True,
    margins_name='All',
    observed=False,
)  
```

```python
HYPER_VRS13 = HYPER_VRS1.pivot_table(
    values=['MEE_BO_VRS_SAME', 'MEE_O_VRS_SAME'],
    index=dmuname,
    columns='year',
    aggfunc='mean',
    fill_value=None,
    margins=False,
    dropna=True,
    margins_name='All',
    observed=False,
)  
```

```python
HYPER_VRS23 = HYPER_VRS2.pivot_table(
    values=['MEE_BO_VRS_DIFF', 'MEE_O_VRS_DIFF'],
    index=dmuname,
    columns='year',
    aggfunc='mean',
    fill_value=None,
    margins=False,
    dropna=True,
    margins_name='All',
    observed=False,
)  
```

## NDDF

```python
from deabook import DEAweak
from deabook.constant import CET_ADDI, RTS_VRS1, RTS_VRS2,RTS_CRS, OPT_DEFAULT, OPT_LOCAL
```

```python
def NDDF_VRS_SAME():

    model = DEAweak.NDDFweak2(data,sent = "K L E = Y:CO2",  gy=[1], gx=[0,0,0],gb=[1],rts=RTS_VRS1, baseindex=None,refindex=None)
    res = model.optimize(solver="mosek")
    # print(model.get_rhoy())
    return res

NDDF_VRS_SAME = NDDF_VRS_SAME()
NDDF_VRS_SAME
```

```python
NDDF_VRS_SAME.rename(columns={'teuo':'EB_VRS_SAME','teo':'EO_VRS_SAME', 'teuo2o':'EBO_VRS_SAME'}, inplace=True)
NDDF_VRS_SAME = NDDF_VRS_SAME[['EB_VRS_SAME','EO_VRS_SAME','EBO_VRS_SAME']]
NDDF_VRS_SAME2 = pd.concat([NDDF_VRS_SAME,yearid],axis=1)
```

```python
NDDF_VRS_SAME3 = NDDF_VRS_SAME2.pivot_table(
    values=['EB_VRS_SAME','EO_VRS_SAME','EBO_VRS_SAME'],
    index=dmuname,
    columns='year',
    aggfunc='mean',
    fill_value=None,
    margins=False,
    dropna=True,
    margins_name='All',
    observed=False,
)  
```
