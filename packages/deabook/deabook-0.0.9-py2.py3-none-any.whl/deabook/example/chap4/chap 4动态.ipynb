## MQDEA

```python
def MQDEA_VRS():

    # model = DEA.DEA(data,sent = "K L E=Y:CO2",  orient="L", rts=RTS_VRS, baseindex=None,refindex=None)
    model = MQDEA.MQDEA(data,id=dmuname,year='year',sent = "K L E=Y",  gx=[0,0,1],gy=[0], rts=RTS_VRS1,\
                        tech = CONTEMPORARY,email = OPT_LOCAL,solver="mosek")
    res = model.optimize()
    # print(model.info(1))
    return res
MQDEA_VRS = MQDEA_VRS()
MQDEA_VRS
```

```python

MQDEA_VRS2 = MQDEA_VRS[['year',dmuname,'MQ',	'MEFFCH',	'MTECHCH']]
```

```python
MQDEA_VRS2.query('year==2017').sort_values('MQ')
```

```python
MQDEA_VRS3 = MQDEA_VRS2.query('year>2016').pivot_table(
    values=['MQ',	'MEFFCH',	'MTECHCH'],
    index=dmuname,
    columns='year',
    aggfunc='mean',
    fill_value=None,
    margins=False,
    dropna=True,
    margins_name='All',
    observed=False,
)    [['MEFFCH','MTECHCH','MQ']]
```
