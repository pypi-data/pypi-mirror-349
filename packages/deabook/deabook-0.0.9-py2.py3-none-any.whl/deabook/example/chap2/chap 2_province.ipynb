## DDF

```python
from deabook import DEA
from deabook.constant import CET_ADDI, RTS_VRS1, RTS_CRS, OPT_LOCAL
```

```python
def DDF_VRS():

    model = DEA.DDF2(data,sent = "K L E=Y",  gy=[0], gx=[1,0,0],rts=RTS_VRS1, baseindex=None,refindex=None)
    res = model.optimize(solver="mosek")
    # print(model.info(1))
    return res

DDF_VRS=DDF_VRS()
DDF_VRS
```

```python
DDF_VRS['EI_VRS'] = DDF_VRS['tei']
DDF_VRS = DDF_VRS.drop(columns = ['optimization_status','rho'])
DDF_VRS2 = pd.concat([DDF_VRS,yearid],axis=1)
```

```python
DDF_VRS3 = DDF_VRS2.pivot_table(
    values=['EI_VRS',],
    index=dmuname,
    columns='year',
    aggfunc='mean',
    fill_value=None,
    margins=False,
    dropna=True,
    margins_name='All',
    observed=False,
)  
```
