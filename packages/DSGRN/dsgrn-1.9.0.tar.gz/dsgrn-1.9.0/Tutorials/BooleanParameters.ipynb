```python
import DSGRN
```

## Boolean parameters

Define a DSGRN network and its parameter graph

```python
net_spec = """x : x + y
              y : (~x) y"""

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network)
```

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('Number of parameter nodes:', parameter_graph.size())
```

Use the flag `B` to define the parameter graph with only Boolean parameters.

```python
net_spec = """x : x + y : B
              y : (~x) y : B"""

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network)
```

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('Number of parameter nodes:', parameter_graph.size())
```

The flag `B` can be combined with the flag `E` to define the parameter graph with the essential Boolean parameters.

```python
net_spec = """x : x + y : B E
              y : (~x) y : B E"""

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network)
```

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('Number of parameter nodes:', parameter_graph.size())
```

## Boolean parameter indices and neighbors

```python
net_spec = """x : x + y : E
              y : (~x) y : E"""

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network)
```

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('Number of parameter nodes:', parameter_graph.size())
```

Find indices of Boolean parameters in the parameter graph of the network above.

```python
bool_parameters = DSGRN.boolean_parameters(parameter_graph)

print(len(bool_parameters))
```

```python
print(bool_parameters)
```

Get the list of nodes in the parameter graph that are neighbors of Boolean parameters.

```python
bool_parameters, bool_par_neighbors = DSGRN.boolean_parameter_neighbors(parameter_graph)

print(len(bool_par_neighbors))
```

## Multi-Boolean parameters

The flag `M` defines a network with a multi-Boolean parameter graph.

```python
net_spec = """x : x + y
              y : (~x) y"""

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network)
```

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('Number of parameter nodes:', parameter_graph.size())
```

```python
net_spec = """x : x + y : M
              y : (~x) y : M"""

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network)
```

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('Number of parameter nodes:', parameter_graph.size())
```

For the network above the DSGRN parameter graph and the multi-Boolean parameter graph are the same. However, for networks with nodes having more than 2 inputs the multi-Boolean parameter graph may have more nodes that the DSGRN parameter graph. The DSGRN parameter graph is always a subgraph.

```python
net_spec = """x : (~y) (~z) + x
              y : ~x
              z : ~y + x"""

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network)
```

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('Number of parameter nodes:', parameter_graph.size())
```

```python
net_spec = """x : (~y) (~z) + x : M
              y : ~x : M
              z : ~y + x : M"""

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network)
```

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('Number of parameter nodes:', parameter_graph.size())
```

The flag `M` can be combined with the flag `E` to define the parameter graph with the essential multi-Boolean parameters.

```python
net_spec = """x : (~y) (~z) + x : M E
              y : ~x : M E
              z : ~y + x : M E"""

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network)
```

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('Number of parameter nodes:', parameter_graph.size())
```
