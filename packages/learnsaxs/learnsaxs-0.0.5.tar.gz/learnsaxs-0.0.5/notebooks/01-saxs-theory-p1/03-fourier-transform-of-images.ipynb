# 3. Fourier Transform of 2D Images
* Inspired by [New Live Fourier Transform code - Brian Richard Pauw](https://lookingatnothing.com/index.php/archives/991)
* Although our goal is the 3D space, let's do some exercise using 2D space. 
* Note here that we don't intend to explain Fourier transform, but instead, try to accept it for granted by observing the results.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
x = y = np.arange(N)
xx, yy = np.meshgrid(x, y)
```

```python
def plot_a_circle_with_ft(a, b, r):
    shape = (xx - a)**2 + (yy - b)**2 < r**2
    canvas = np.zeros((N,N))
    canvas[shape] = 1
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8,4))
    ax1.imshow(canvas)
    ft_image = np.abs(np.fft.fft2(canvas))
    ax2.imshow(ft_image)
```

```python
plot_a_circle_with_ft(50, 50, 10)
```

```python
from ipywidgets import interactive
interactive_plot = interactive(plot_a_circle_with_ft, a=(0,99), b=(0,99), r=(0,10))
interactive_plot
```
